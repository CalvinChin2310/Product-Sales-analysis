# Product Sales Analysis


#### importing libraries

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

#### Merging all 12 month data into one CSV

```python
path = "./Product_Sales_Data"
files = [file for file in os.listdir(path) if not file.startswith('.')] # Ignore hidden files

all_months_data = pd.DataFrame()

for file in files:
    current_data = pd.read_csv(path+"/"+file)
    all_months_data = pd.concat([all_months_data, current_data])
    
all_months_data.to_csv("all_data_copy.csv", index=False)
```

#### Read in updated dataframe

```python
all_data = pd.read_csv("all_data_copy.csv")
all_data.head()
```

### Data Cleaning

In the process of data cleaning, the initial step involves identifying areas that require attention. Typically, these areas become apparent as you execute operations and encounter errors. Upon encountering an error, you evaluate the data to determine the appropriate cleaning actions needed to resolve the issue.

##### Drop rows of NAN

```python
# Finding NAN values in the dataset
nan_df = all_data[all_data.isna().any(axis=1)]
display(nan_df.head())

all_data = all_data.dropna(how='all')
all_data.head()
```

##### Remove any textual content from the 'Order Date' column.

###### Reasons for this is because:
Consistency: Keeping the 'Order Date' column consistent ensures that all values are in a standardized format, making it easier to analyze and compare the dates.

Analysis: Textual content in the 'Order Date' column can hinder analysis, especially when performing date-based operations such as sorting, grouping, or calculating time intervals.

Data Integrity: Removing unnecessary text helps maintain the integrity of the dataset by ensuring that each entry in the 'Order Date' column represents a valid date format, reducing the risk of errors or inconsistencies in downstream analyses.

Ease of Use: Clean data is more user-friendly and can be more readily utilized by other team members or systems for further analysis or reporting purposes.

```python
all_data = all_data[all_data['Order Date'].str[0:2]!='Or']
```

#### Ensure that the columns have the appropriate data types.

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

### Augment data with additional columns

#### Adding a new "Month" column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

#### Adding a new "City" column

```python
def get_city(address):
    return address.split(",")[1].strip(" ")

def get_state(address):
    return address.split(",")[2].split(" ")[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)}  ({get_state(x)})")
all_data.head()
```

```python
all_data['Sales'] = all_data['Quantity Ordered'].astype('int') * all_data['Price Each'].astype('float')
```

#### Question 3: What time should we display advertisements to maximize likelihood of customer's buying product?

```python
# Add Hour, Minute, Count column
#all_data['Hour'] = pd.to_datetime(all_data['Order Date']).dt.hour.This line creates a new column named 'Hour' in the DataFrame all_data. 
#It converts the 'Order Date' column to datetime format using pd.to_datetime().
#Then, it extracts the hour component from the datetime using the .dt.hour accessor.
#This will populate the 'Hour' column with the hour values (0-23) corresponding to the timestamps in the 'Order Date' column.
#all_data['Minute'] = pd.to_datetime(all_data['Order Date']).dt.minute

#Similar to the previous line, this creates a new column named 'Minute'.
#It converts the 'Order Date' column to datetime format.
#Then, it extracts the minute component from the datetime using the .dt.minute accessor.
#This will populate the 'Minute' column with the minute values (0-59) corresponding to the timestamps in the 'Order Date' column.
#all_data['Count'] = 1

#This line creates a new column named 'Count' in the DataFrame all_data.
#It assigns a constant value of 1 to each row in the 'Count' column.
#This is often used for counting or aggregation purposes. Each row having a value of 1 can be later summed up to get counts.

all_data['Hour'] = pd.to_datetime(all_data['Order Date'], format='%m/%d/%y %H:%M').dt.hour
all_data['Minute'] = pd.to_datetime(all_data['Order Date'], format='%m/%d/%y %H:%M').dt.minute
all_data['Count'] = 1
all_data.head()
```

```python
keys = list(all_data.groupby(['Hour']).groups.keys())

plt.plot(keys, all_data.groupby(['Hour']).count()['Count'])
plt.xticks(keys)
plt.grid()
plt.show()


```

#### Question 4: What products are most often sold together?

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)].copy()

# Use .copy() to ensure that modifications are made on a copy of the DataFrame

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

# Use .loc[] to set values
df.loc[:, 'Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df2 = df[['Order ID', 'Grouped']].drop_duplicates()
```

```python
# Referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
from itertools import combinations
from collections import Counter

count = Counter()

for row in df2['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    print(key, value)
```

#### Based on the value above what products do you think sold the most and why?

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']

keys = [pair for pair, df in product_group]
plt.bar(keys, quantity_ordered)
plt.xticks(keys, rotation='vertical', size=8)
plt.show()
```

```python
# Convert 'Price Each' column to numeric
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'], errors='coerce')

# Group by 'Product' and calculate the mean of 'Price Each'
prices = all_data.groupby('Product')['Price Each'].mean()

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(keys, quantity_ordered, color='g')
ax2.plot(keys, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')

# Set the x-axis ticks and labels
ax1.set_xticks(range(len(keys)))
ax1.set_xticklabels(keys, rotation='vertical', size=8)

plt.show()

```
